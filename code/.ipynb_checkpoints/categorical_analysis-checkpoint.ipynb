# Distribution if using classical mode

```python
mode_count_dic = {}
master_list = []

mode_dic = {}

for this_key in article_all.keys():
    
    responses = article_all[this_key]
    
    responses = responses[:6]
    none_flag = (responses[-1] == None)
    
    responses = [str(x) for x in responses]
    
    counts = Counter(responses)
        
    nums_count_list = []
    
    for keys_count, nums_count in counts.items():
    
        nums_count_list.append(nums_count)
        
    max_count = np.max(nums_count_list)
    
    out_put_list = []
    
    for keys_count, nums_count in counts.items():
            
            if nums_count == max_count:
                
                out_put_list.append(keys_count)
                
    out_put_list.sort()
    
    master_list.append(tuple(out_put_list))
    mode_count_dic[this_key] = out_put_list
    
    
    cur_list = mode_dic.get(tuple(out_put_list),[])
    
    cur_list.append(this_key)
    
    mode_dic[tuple(out_put_list)] = cur_list
    
    
```

```python
# Graphing
tally_mode = Counter(master_list)
keys_mode = tally_mode.keys()
keys_mode = [str(x) for x in keys_mode]
values_mode = tally_mode.values()

plt.title("Pure Mode")
plt.bar(keys_mode, values_mode)
```

## What do each articles have?

```python
all_list = []

overall_dic = {}

for this_key in article_all.keys():
    
    responses = article_all[this_key]
    
    responses = responses[:6]
    
    while None in responses:
        
        responses = responses[:len(responses)-1]
    
    
    none_flag = (responses[-1] == None)
    
    responses = [str(x) for x in responses]
    
    
    
    resp_count = Counter(responses)
    
    
    
    resp_count = list(resp_count.keys())
    
    resp_count.sort()
    
    all_list.append(tuple(resp_count))

    cur_list = overall_dic.get(tuple(resp_count),[])

    cur_list.append(this_key)

    overall_dic[tuple(resp_count)] = cur_list
```

```python
# Graphing
tally_all = Counter(all_list)
keys_all = tally_all.keys()
keys_all = [str(x) for x in keys_all]
values_all = tally_all.values()

plt.title("Overall Distribution")
plt.bar(keys_all, values_all)
plt.xticks(rotation=45)
```

## Leave one out analysis

```python
out_list = []
leave_dic = {} #this list stores which articles went to what.

for this_key in article_all.keys():
    
    responses = article_all[this_key]
    
    responses = responses[:6]
    
    while None in responses:
        
        responses = responses[:len(responses)-1]
        
    resp_count_out = Counter(responses)
    
    resp_array = np.array(list(resp_count_out.values()))
    
    if len(resp_array) != 1:
    
        resp_array.sort()
                          
    
    if len(resp_array) == 1:
        
        out_list.append(list(resp_count_out.keys())[0])
        
        cur_list = leave_dic.get(list(resp_count_out.keys())[0], [])
        
        cur_list.append(this_key)
        
        leave_dic[list(resp_count_out.keys())[0]] = cur_list
        
    
    elif resp_array[-1] - 1 <= resp_array[-2]:
        
        out_list.append("S")
        
        cur_list = leave_dic.get("S", [])
        
        cur_list.append(this_key)
        
        leave_dic["S"] = cur_list
        
    else:
        
        for keys_count, nums_count in resp_count_out.items():
            
            if nums_count == np.max(resp_array):
                
                out_list.append(keys_count)
                
                cur_list = leave_dic.get(keys_count, [])
                
                cur_list.append(this_key)
                
                leave_dic[keys_count] = cur_list
        
    
        
    
    
```

```python
out_list

# Graphing
tally_out = Counter(out_list)
keys_out = tally_out.keys()
keys_out = [str(x) for x in keys_out]
values_out = tally_out.values()

plt.title("Leave One Out")
plt.bar(keys_out, values_out)
#plt.xticks(rotation=45)
```

# Overlap Analysis

```python
#leave one out false
false_leave_one_out = set(leave_dic['f'])

#mode false
false_mode = set(mode_dic[('f',)])

#no true
copy_list = overall_dic[('c', 'f')].copy()
copy_list.extend(overall_dic[('f',)])
false_no_true = set(copy_list)
```

```python
# How many articles are not in leave one out and pure mode?
false_leave_one_out.symmetric_difference(false_mode)
```

```python
# How many articles are not in leave one and False no true
len(false_leave_one_out.symmetric_difference(false_no_true))
```

```python
false_sets = [false_leave_one_out,false_mode,false_no_true]
false_labels = ("Robust Mode", "Mode", "No True")
plt.figure(figsize=(7.5,7.5))

plt.title("False Articles", size = 15)
v = venn3(subsets = false_sets, set_labels = false_labels)


h = list(v.patches)
h = [x for x in h if x != None]
false_labels = ["Mode",
                "Mode + Robust Mode",
                "No True",
                "Mode + No True",
                "Mode + Robust Mode + No True"]
ax = plt.gca()
ax.legend(handles=h, labels=false_labels,  bbox_to_anchor=(1, 1), fontsize='large')
plt.savefig(my_figs + "/venn_false.png", dpi = 100, bbox_inches='tight')
```

```python
#the super set of all 
super_set_false = false_mode.copy()
super_set_false = super_set_false.union(false_no_true)

#the high confidence false
high_con_false = false_leave_one_out.copy()
high_con_false = high_con_false.intersection(false_no_true)
```

# Overlap for True

```python
#leave one out false
true_leave_one_out = set(leave_dic['t'])

#mode false
true_mode = set(mode_dic[('t',)])

#no true
copy_list_true = overall_dic[('c', 't')].copy()
copy_list_true.extend(overall_dic[('t',)])
true_no_false = set(copy_list_true)
```

```python
plt.figure(figsize=(7.5,7.5))

true_sets = [true_leave_one_out,true_mode,true_no_false]
true_labels = ("Robust Mode", "Mode", "No False")
plt.title("True Articles", size = 15)
venn3(true_sets, true_labels)


h = list(v.patches)
h = [x for x in h if x != None]
false_labels = ["Mode",
                "Mode + Robust Mode",
                "No True",
                "Mode + No False",
                "Mode + Robust Mode + No False"]
ax = plt.gca()
ax.legend(handles=h, labels=false_labels,  bbox_to_anchor=(1, 1), fontsize='large')
plt.savefig(my_figs + "/venn_true.png", dpi = 100, bbox_inches='tight')
```

```python
#the super set of all 
super_set_true = true_mode.copy()
super_set_true = super_set_true.union(true_no_false)

#the high confidence false
high_con_true = true_leave_one_out.copy()
high_con_true = high_con_true.intersection(true_no_false)
```

# Creating data on these for kevin

```python
#  article_dic
# 1: ('CND', 2.17, 'nov 13', 1)

#format

#date, type, macro_number, mode, robust_mode(yes or no or None), no_contradiction_true, all_true, any_true
# no_contradiction_false, any_false, all_false

all_list = []

for this_key in article_all.keys():
    
    mode_est, _ , date, art_type = article_dic[this_key]
    
    responses = article_all[this_key]
    
    responses = responses[:6]
    
    while None in responses:
        
        responses = responses[:len(responses)-1]
    
    resp_count_out = Counter(responses)
    
    common = resp_count_out.most_common()
    
    if mode_est == "CND" or mode_est == "No Mode!":
        
        robust_mode = None
        
    else:
        
        robust_mode = (this_key in false_leave_one_out or this_key in true_leave_one_out)
        
    
    no_contradiction_true = (this_key in true_no_false)
    all_true = (this_key in high_con_true)
    any_true = (this_key in super_set_true)
    
    no_contradiction_false = (this_key in false_no_true)
    all_false = (this_key in high_con_false)
    any_false = (this_key in super_set_false)
    
    output_list = [date, art_type, this_key, mode_est, robust_mode]
    output_list_true = [no_contradiction_true,all_true,any_true]
    output_list_false = [no_contradiction_false, all_false, any_false]
    
    output_list.extend(output_list_true)
    output_list.extend(output_list_false)
    
    all_list.append(output_list)
    
col_list = ['date', 'article_type',"article_number","mode","in_robust_mode","no_false", \
           "all_true_set", "any_true_set","no_true","all_false_set","any_false_set"]
all_data = pd.DataFrame(all_list, columns = col_list)
```
